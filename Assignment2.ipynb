```python
class Node:
    """Class representing a node in the singly linked list."""
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    """Class representing the singly linked list."""
    def __init__(self):
        self.head = None

    def add_node(self, data):
        """Adds a node with the given data to the end of the list."""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        # Traverse to the last node
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def print_list(self):
        """Prints the contents of the list."""
        if not self.head:
            print("The list is empty.")
            return
        current = self.head
        print("Linked List:", end=" ")
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    def delete_nth_node(self, n):
        """
        Deletes the nth node (1-based index) from the list.
        Raises an exception for invalid operations.
        """
        if not self.head:
            raise Exception("Cannot delete from an empty list.")

        if n <= 0:
            raise ValueError("Index must be 1 or greater.")

        # If deleting the head node
        if n == 1:
            deleted_data = self.head.data
            self.head = self.head.next
            print(f"Deleted node at index {n} with data: {deleted_data}")
            return

        current = self.head
        prev = None
        count = 1

        while current and count < n:
            prev = current
            current = current.next
            count += 1

        if not current:
            raise IndexError(f"Index {n} is out of range.")

        prev.next = current.next
        print(f"Deleted node at index {n} with data: {current.data}")


# === Sample Test Case ===
if __name__ == "__main__":
    # Create a linked list
    ll = LinkedList()

    # Add sample nodes
    ll.add_node(10)
    ll.add_node(20)
    ll.add_node(30)
    ll.add_node(40)
    ll.print_list()  # Output: 10 -> 20 -> 30 -> 40 -> None

    # Delete the 2nd node
    try:
        ll.delete_nth_node(2)  # Should delete 20
    except Exception as e:
        print("Error:", e)
    ll.print_list()  # Output: 10 -> 30 -> 40 -> None

    # Try deleting a node with index out of range
    try:
        ll.delete_nth_node(10)
    except Exception as e:
        print("Error:", e)

    # Delete all nodes one by one
    try:
        ll.delete_nth_node(1)
        ll.delete_nth_node(1)
        ll.delete_nth_node(1)
    except Exception as e:
        print("Error:", e)

    ll.print_list()  # Should print: The list is empty.

    # Try deleting from an empty list
    try:
        ll.delete_nth_node(1)
    except Exception as e:
        print("Error:", e)
```
